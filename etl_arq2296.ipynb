```python
# Função para extração de data e hora do nome do arquivo
def get_datetime(file):
    data = file[8:16] + file[17:-4]
    return(data)
```

```python
def norm_2296 (files2296):
    for arq in files2296:
        df = pd.read_fwf( arq, header=None, encoding='cp1252', 
                     colspecs=cols_specs, names= columns_2296)
        df['data'] = get_datetime(arq) # cria coluna data
        df['nome_arquivo'] = arq # cria coluna arquivo
        # criar df final com todos os dados
        if 'df_final' not in locals():
            df_final = df
        else:
            df_final = pd.concat([df_final, df])
        print(f"Processando arquivo: {arq}")
        # salvar df final em csv
        df_final.to_csv(r'G:\Outros computadores\acer\gits\ptec009_automacao_indicadores_processorio\dados_transformados\etl_arqs\expedientes_arqs2296.csv', index=False, sep=';')
    return(df_final)
    # # salvar df em um arquivo csv
    # df_final.to_csv(path_arqs + '/arqs2296/' + 'dados_2296.csv' , index=False, sep=';')
    
norm_2296(files2296)
```
