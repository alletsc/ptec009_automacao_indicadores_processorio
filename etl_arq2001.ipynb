```python
import pandas as pd
import os
import glob
```

```python
# definindo o caminho dos arquivos
path_arqs = 'G:/Outros computadores/acer/gits/ptec009_automacao_indicadores_processorio/dados_originais/rede/PROCESSORIO_SICOP'
os.chdir(path_arqs)
```

```python
# separando os arquivos em pastas

def selec_arqs(path_arqs):
    for arq in os.listdir():
        if 'ARQ2001_' in arq:
            # mover para pasta arqs2001
            os.rename(arq, path_arqs + r'/arqs2001/'+ arq)
        elif 'ARQ2296' in arq:
            #mover para pasta arqs2296
            os.rename(arq, path_arqs + r'/arqs2296/' + arq)
```

```python
# aplicando a função
selec_arqs(path_arqs)
```

```python
# Listagem dos arquivos na pasta arqs2001
dir2001 = path_arqs + r'/arqs2001/'
files2001 = os.listdir(dir2001)
```

```python
files2001
```

```python
# Colunas dos arquivos 2001

columns_2001 = ['orgao_transcritor', 'codigo_sici', 'numero_processo', 'tipo_processo', 'numero_documento_ident', 'tipo_documento_ident', 'descricao_tipo_documento_ident',
                'requerente', 'data_processo', 'codigo_assunto', 'descricao_assunto', 'data_cadastro', 'assunto_complementar', 'matricula_digitador', 'prazo_cadastro', 'arquivo_origem', 'data'] 
```

```python
def norm_line(line):

    split_line = line.split(';') # Separa a linha por ";" e cria uma lista 
    n = len(split_line) # Número de itens da lista

    new_line = split_line[:12] # Primeiros 12 itens da lista
    new_line.append(' |'.join(split_line[12:n -2])) # Adição dos itens variáveis concatenados por "|" 
    new_line.extend(split_line[-2:]) # Adição dos itens finais da lista
    new_line = [x.strip() for x in new_line] # Remoção de espaços em branco
    return(new_line) # Retona a nova linha

# Função para extração de data e hora do nome do arquivo

def get_datetime(file):
    data = file[8:16] + file[17:-4]
    return(data)
```

```python
with open(dir2001 + 'dados_2001.txt', 'w') as output_file: # Cria um arquivo 'dados_2001.txt' no diretório e deixa ele aberto
    for file in files2001: 
        # Para cada arquivo existente no diretório arqs2001
        print(f'Processando arquivo {file}')
        with open(dir2001 + file, 'r') as input_file: # Abre o arquivo somente leitura
            lines = input_file.readlines() # Gera uma lista contendo todas as linhas do arquivo
            for line in lines: # Para cada linha da lista
                new_list = norm_line(line) # Aplica função de normalização gerando uma nova lista
                new_list.append(file) # Adição do nome do arquivo
                new_list.append(get_datetime(file)) # Adição da data e hora do arquivo
                new_line = ';'.join(new_list) # Concatenação dos itens com ";"
                output_file.write(new_line + '\n') # Adiciona a nova linha ao arquivo 'compilado_2001.txt'
```

```python
# ler arquivo gerado

df = pd.read_csv(dir2001 + 'dados_2001.txt', sep=';', encoding = 'cp1251', header = None, dtype = str) # Leitura do arquivo 'compilado_2001.txt'
df.columns = columns_2001 # Adição das colunas
df.head()
```

```python
# formatar coluna data como date_time
df['data'] = pd.to_datetime(df['data'])
```

```python
# salvar como arquivo csv separado por na pasta dados transformados;
df.to_csv(r'G:\Outros computadores\acer\gits\ptec009_automacao_indicadores_processorio\dados_transformados\etl_arqs\processos_arqs2001.csv', index = False, sep=';')
# apagar dados.txt
os.remove(dir2001 + 'dados_2001.txt')
```
